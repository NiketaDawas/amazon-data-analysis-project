# Data cleaning

```python
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(float)
#convert string type data in float type becouse can't be grouby with str
```

```python
df['rating_count'] = df.groupby('category')['rating_count'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
#to convert the string type data in numerical and remove the ₹, ',' sign
df['actual_price'] = df['actual_price'].astype(str)  # convert to string
df['actual_price']=df['actual_price'].str.replace(',','')
df['actual_price']=df['actual_price'].str.replace('₹','')
df['actual_price']=df['actual_price'].astype(float)
```

```python
df['discounted_price'] = df['discounted_price'].astype(str)  # convert to string
df['discounted_price']=df['discounted_price'].str.replace('₹','')
df['discounted_price']=df['discounted_price'].str.replace(',','')
df['discounted_price']=df['discounted_price'].astype(float)
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str)  # convert to string

df['discount_percentage'] = df['discount_percentage'].str.replace('%','')
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df['rating_count'] = df['rating_count'].astype(str)  # convert to string

df['rating_count'] = df['rating_count'].str.replace(',','')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

# Data visulaization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure()
df['rating'].hist()
plt.title("Rating distribution")
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

Most of the products have ratings between 4 and 5, which indicates high customer satisfaction on Amazon

```python
top_rated = df.sort_values(by='rating', ascending=False).head(10)
```

```python
plt.figure(figsize=(40,40))
plt.barh(top_rated['product_name'],top_rated['rating'])
plt.xlabel('Rating')
plt.ylabel('Products')
plt.title("top 10 rating product name")
plt.yticks(fontsize=52)   # product name readable banane ke liye
plt.show()
```

The bar chart highlights the top 10 highest rated products, showing that these items deliver excellent customer satisfaction

```python
top_discount = df.sort_values(by='discount_percentage', ascending=False).head(10)

plt.figure()
plt.barh(top_discount['product_name'], top_discount['discount_percentage'])
plt.xlabel("Discount %")
plt.ylabel("Product")
plt.title("Top 10 Discounted Products")
plt.show()
```

```python
cat_count = df['category'].value_counts().head(10)

plt.figure()
cat_count.plot(kind='bar')
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.title("Top 10 Categories by Product Count")
plt.show()
```

```python
sample = df.head(20)

plt.figure()
plt.plot(sample['actual_price'], label='Actual Price')
plt.plot(sample['discounted_price'], label='Discounted Price')
plt.legend()
plt.title("Actual vs Discounted Price")
plt.show()
```

##Since the discounted price line consistently lies below the actual price line, it indicates that significant discounts are applied across most products

```python
plt.figure()
plt.scatter(df['actual_price'], df['rating'])
plt.xlabel("Actual Price")
plt.ylabel("Rating")
plt.title("Price vs Rating")
plt.show()
```

The scatter plot shows a scattered pattern, indicating that product price does not significantly affect customer ratings.
